```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
#encode the necessary columns
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

le = LabelEncoder()
df['EducationBackground'] = le.fit_transform(df['EducationBackground'])

le = LabelEncoder()
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])

le = LabelEncoder()
df['EmpDepartment'] = le.fit_transform(df['EmpDepartment'])

le = LabelEncoder()
df['EmpJobRole'] = le.fit_transform(df['EmpJobRole'])

le = LabelEncoder()
df['BusinessTravelFrequency'] = le.fit_transform(df['BusinessTravelFrequency'])

le = LabelEncoder()
df['Attrition'] = le.fit_transform(df['Attrition'])

le = LabelEncoder()
df['EmpNumber'] = le.fit_transform(df['EmpNumber'])

le = LabelEncoder()
df['OverTime'] = le.fit_transform(df['OverTime'])
```

```python
#confirm the columns are encoded by viewing the first five rows
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#prepare my data for testing and training
x = df.drop('PerformanceRating' , axis=1)
y = df['PerformanceRating']
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
#choose a suitable predictive model


#Regression Models: Linear Regression, Random Forest Regression,
#Gradient Boosting Regression, Support Vector Regression (SVR).


#Classification Models: Logistic Regression, Random Forest Classifier,
#Gradient Boosting Classifier, Support Vector Machine (SVM), k-Nearest Neighbors (k-NN).


#evaluate the models by training different models and
#check their performance using mean squared error, r2 score and Accuracy.
#Choose the model that performs best on my evaluation metrics.
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, classification_report, accuracy_score, r2_score
from sklearn.neighbors import KNeighborsClassifier
```

***KNN***

```python
#create KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
#fit the classifier into the training dataset
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
#print the accuracy
print("Accuracy:",knn.score(x_test,y_test))
```

***Random Forest Classifier***

```python
#create and fit the model
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(x_train,y_train)
```

```python
#make predictions
y_pred = rfc.predict(x_test)
```

```python
#check classification report
print('Classification Report', classification_report(y_test,y_pred))
```

```python
#check accuracy
print('Accuracy', accuracy_score(y_test,y_pred))
```

***Tune & Boost My Model***

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
#Initialize the GridSearchCV object
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42), param_grid=param_grid, cv=5, n_jobs=-1)
```

```python
#Fit the GridSearchCV object to the data
grid_search.fit(x_train, y_train)
```

```python
#Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
#Use the best estimator to make predictions
best_rf_model = grid_search.best_estimator_
y_pred = best_rf_model.predict(x_test)
```

```python
#Evaluate the model
print('Best Parameters:', best_params)
print('Best Score:', best_score)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Classification Report:', classification_report(y_test, y_pred))
```
